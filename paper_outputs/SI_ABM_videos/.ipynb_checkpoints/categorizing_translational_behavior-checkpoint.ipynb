```python
#define some functions

#maybe we can find omega by taking the derivatives
def omega_counter(x_array, y_array):
    dy = [y_array[ii]-y_array[ii-1] for ii in range(len(fit_y_data_herd))]
    dx = [x_array[ii]-x_array[ii-1] for ii in range(len(fit_y_data_herd))]
    derivative = dy #[dy[ii]/dx[ii] for ii in range(len(fit_y_data_herd))]

    #calculate omega 
    omega_counter = 0
    for ii in range(len(derivative)):
        if derivative[ii-1] >0:
            if derivative[ii] <0:
                omega_counter+=1
        if derivative[ii-1] <0:
            if derivative[ii] >0:
                omega_counter+=1
    print(omega_counter)
    
    return omega_counter
```
