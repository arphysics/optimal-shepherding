## Now exploring fit parameters

```python
from scipy.optimize import curve_fit

def omega_double(x, a, b):
    
    return a*1.0/(0.1+b*x)


def omega_single(x, a):
    
    return a*1.0/(0.1+1.0*x)
```

```python
popt_double_drove, pcov_double_drove = curve_fit(omega_double, temp_scale_x, temp_scale_peak_d)
popt_double_must, pcov_double_must = curve_fit(omega_double, temp_scale_x, temp_scale_peak)
popt_single_drove, pcov_single_drove = curve_fit(omega_single, temp_scale_x, temp_scale_peak_d)
popt_single_must, pcov_single_must = curve_fit(omega_single, temp_scale_x, temp_scale_peak)
```
